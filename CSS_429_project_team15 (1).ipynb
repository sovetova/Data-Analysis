```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.graph_objs as go
import plotly.offline as py
```

```
data_math = pd.read_csv("student-mat.csv")
```

```
data_por = pd.read_csv("student-por.csv")
```

# Vizualization

```
sns.catplot(x="sex", kind="count",palette="viridis", data=data_math, height = 6)
plt.title("Gender of students : F - female,M - male")
```

```
ages = data_math["age"].value_counts()
labels = (np.array(ages.index))
sizes = (np.array((ages / ages.sum())*100))
trace = go.Pie(labels=labels, values=sizes)
layout = go.Layout(title="Аge of students")
dat = [trace]
fig = go.Figure(data=dat, layout=layout)
py.iplot(fig, filename="age")
```

```
sns.catplot(x="higher", kind="count",palette="flare", data=data_math, height = 3)
plt.title("The desire to get higher education.")
```

```
plt.figure(figsize=(10,3))
plt.title("Box plot for final grades,depending on the desire to have higher education")
sns.boxplot(y="higher", x="G3", data = data_math , orient="h", palette = 'viridis')
```

```
plt.figure(figsize=(10,3))
plt.title("Box plot for final grades,depending on mothers education")
sns.boxplot(y="Medu", x="G3", data = data_math , orient="h", palette = 'magma')

plt.figure(figsize=(10,3))
plt.title("Box plot for final grades,depending on fathers education")
sns.boxplot(y="Fedu", x="G3", data = data_math , orient="h", palette = 'viridis')
```

```
data_math['st_time'] = np.nan
df = [data_math]
for col in df:
    col.loc[col['studytime'] == 1 , 'st_time'] = '< 2 hours'
    col.loc[col['studytime'] == 2 , 'st_time'] = '2 to 5 hours'
    col.loc[col['studytime'] == 3, 'st_time'] = '5 to 10 hours'
    col.loc[col['studytime'] == 4, 'st_time'] = '> 10 hours'  
plt.figure(figsize=(10,3))
plt.title("Box plot for final grades,depending on the study time")
sns.boxplot(y="st_time", x="G3", data = data_math , orient="h", palette = 'flare')
```

```
plt.figure(figsize=(10,3))
plt.title("Box plot for final grades,depending on mothers profession")
sns.boxplot(y="Medu", x="G3", data = data_math , orient="h", palette = 'magma')
```
